```python
# Python program to calculate the sum of elements in a list using sum()

numbers = []
n = int(input("Enter the number of elements: "))
for i in range(n):
  num = float(input("Enter the element: "))
  numbers.append(num)
result = sum(numbers)
print("Sum of elements is ", result)
```

```python
list1 = []
list2 = []
n = int(input("Enter the number of elements: "))

print("Enter the elements for list1: ")
for i in range(n):
  num = float(input("Enter the element: "))
  list1.append(num)

print("Enter the elements for list2: ")
for i in range(n):
  num = float(input("Enter the element: "))
  list2.append(num)

result = list(zip(list1, list2))
print("Combined lists : ", result)
```
