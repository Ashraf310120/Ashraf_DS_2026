```python
weight_kg, height_m = 70, 1.75
bmi = weight_kg / (height_m ** 2)
if bmi < 18.5:
  cat = "Underweight"
elif bmi < 25:
  cat = "Normal"
elif bmi < 30:
  cat = "Overweight"
else:
  cat = "Obese"
print(f"BMI {bmi:.1f} : {cat}")
```
