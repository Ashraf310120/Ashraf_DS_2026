### ---ARTHEMETIC OPERATORS---

```python
### 1. Arithmetic Operators on two numbers

num_1 = float(input("Enter first number: "))
num_2 = float(input("Enter second number: "))

### -----Arithemetic Operators-----

print("Addition: ", num_1 + num_2)
print("Subtraction: ", num_1 - num_2)
print("Multiplication: ", num_1 * num_2)
print("Division: ", num_1 / num_2)
print("Floor Division: ", num_1 // num_2)
print("Modulus: ", num_1 % num_2)
print("Exponentiation: ", num_1 ** num_2)
```

```python
###2. Swaping two numbers

a = 10
b = 5

a, b = b, a     ### Swaps the values of a and b

print("a =", a)
print("b =", b)
```

```python
### 3. Even or odd using modulus operator

number = 20

if number % 2 == 0:   ### num % 2 == 0 returns if it is even
  print("Even number: ", number)
else:
  print("Odd number: ", number)
```

```python
### 4. Reverse a three digit number

num = 143

### rev returns the value in reverse order
rev = (num % 10)  * 100 + ((num // 10) % 10) * 10 + (num // 100)

print("Reverse of the number: ", rev)
```

```python
### 5. Last digit and 1st digit of number

num = 12345

last_digit = num % 10
first_digit = num // 10000

print("Last Digit: ", last_digit)
print("First Digit: ", first_digit)
```

```python
num = int(input("Enter a number: "))

last_digit = num % 10

while num >= 10:
  num //= 10
first_digit = num

print("Last Digit: ", last_digit)
print("First Digit: ", first_digit)
```

```python
### 6. Demonstrate all Assignment Operators (=, +=, -=, *=, /=, //=, %=, **=)

a = 10
print("Initial Value", a)

a+= 10
print(" +=: ", a)

a -= 5
print(" -=: ", a)

a *= 2
print(" *=: ", a)

a /= 3
print(" /=: ", a)

a //= 2
print(" //=: ", a)

a %= 5
print(" %=: ", a)

a **= 7
print(" **= : ", a)
```

### --- Bitwise Operators ---

```python
### 17. Bitwise of AND, OR, XOR of two numbers

a = 10
b = 6
print("Bitwise AND: ", a & b)
print("Bitwise OR: ", a | b)
print("Bitwise XOR: ", a ^ b)
```

```python
### 18. Left Shift and right shift

a = 10
print("Left Shift: ", a << 2)
print("Right Shift: ", a >> 2)
```

```python
### 19. Number is using bitwise or not

num = 10

if num & 1 == 0:           ## num & 1 == 0 >> Even
  print("Number is Even")  ## num & 1 == 1 >> Odd
else:
  print("Number is Odd")
```

```python
### 20. Swap two numbers using XOR

a = 10
b = 20

## Swapping property of XOR

a = a ^ b
b = a ^ b
a = a ^ b

print("After swapping: ")
print("a = ", a)
print("b = ", b)
```

```python
### 21. Count the number of Set Bits(1s)

num = 12

count = 0

while num > 0:
  count += num & 1
  num = num >> 1

print("Number of Set Bits: ", count)
```

### --- Identity Operators ---

```python
### 22. The difference between is and ==

a = [1, 2, 3]
b = [1, 2, 3]
c = a

print("a == b: ", a == b)  ## == compares the value of different objects
print("a is b: ", a is b)  ## is compares variables refer to same object

print("a == c; ", a == c)
print("a is c: ", a is c)
```

```python
### 23. Two variables refer to same list object

a = [1, 2, 3]
b = a

if a is b:
  print("Both variables refer to same list object")
else:
  print("Both variables refer to different list object")
```

```python
### 27. Counting vowels in a string in predefined vowel list using in

word = "Edukron"

vowels = "aeiouAEIOU"
count = 0

for ch in word:
  if ch in vowels: ## checks whether character is vowel
    count += 1    ## count += 1 count is increased by 1

print(word)
print("Number of vowels: ", count)
```

```python
### 28. Simple calculator using Arithmetic and Comparison operators

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

# Arithmetic Operators
print("Addition: ", a + b)
print("Subtraction: ", a - b)
print("Multiplication: ", a * b)
print("Division: ", a / b)

# Comparison Operators
print("a == b: ", a == b)
print("a != b: ", a != b)
print("a > b: ", a > b)
print("a < b: ", a < b)
print("a >= b: ", a >= b)
print("a <= b: ", a <= b)
```

```python
### 29. Grading System using Comparison & Logical Operators

marks = int(input("Enter marks: "))

## Using 'if - elif - else' condition

if marks >= 90 and marks <= 100:
  print("Grade : A ")
elif marks >= 75 and marks < 90:
  print("Grade : B ")
elif marks >= 60 and marks < 75:
  print("Grade : C ")
elif marks >= 45 and marks < 60:
  print("Grade : D ")
elif marks >= 33 and marks < 45:
  print("Grade : E ")
elif marks >= 0 and marks < 33:
  print("Grade : F ")
else:
  print("Invalid marks")
```
