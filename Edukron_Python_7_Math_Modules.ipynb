```python
### 24. Checking the number is finite, infinite, or NaN

number = 25

### math.isfinite() returns True if the number is neither infinity nor NaN
### math.isinf() returns True if the number is positive or negative infinity
### math.isnan() returns True if the number is NaN (Not a number)

print("Is Finite: ", math.isfinite(number))
print("Is Infinite: ", math.isinf(number))
print("Is NaN: ", math.isnan(number))
```

```python
### 25. Product of all numbers in a list

numbers = [1, 2, 3, 4, 5, 6]

### math.prod() calculates the product of all numbers in a list
product = math.prod(numbers)

print("Product:", product)
```

```python
### 26. Sum of square roots of numbers in a list

numbers = [4, 9, 16, 25]

### math.sqrt() calculates the sum of square roots of numbers in a list
sum_of_square_roots = sum(math.sqrt(number) for number in numbers)

print("Sum of Square Roots: ", sum_of_square_roots)
```

```python
### 27. Finding the largest integer <= square root of given number

number = 32

### math.isqrt() returns the integer square root

largest_integer = math.isqrt(number)

print("Largest Integer: ", largest_integer)
```

```python
### 28. Compound intrest using math.pow()

principle = 10000
rate = 5
time = 2

### Using math.pow() to calculate compound intrest

amount = principle * math.pow(1+rate/100, time)

compound_intrest = amount - principle

print("Amount: ", amount)
print("Compound Intrest: ", compound_intrest)
```

```python
### 29. Finding area, circumference of circle using math.pi




radius = 5

###  math.pi provides the value of pi

area = math.pi * radius**2             ### AREA OF CIRCLE = PI*R**2
circumference = 2 * math.pi * radius   ### CIRCUMFERENCE OF CIRCLE = 2*PI*R

print("Area: ", area)
print("Circumference: ", circumference)
```
